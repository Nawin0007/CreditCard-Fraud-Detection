```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('creditcard.csv')

```

```python
# View first few rows
print(df.head())
```

```python

# Basic info and statistics
print(df.info())
print(df.describe())
```

```python

# Check for null values
print(df.isnull().sum())
```

```python

# Data selection and filtering
print(df['Amount'].head())  # Selecting a column
print(df[df['Amount'] > 1000])  # Filtering rows
```

```python

# Grouping and aggregation
print(df.groupby('Class')['Amount'].mean())

# Adding new column
df['Log_Amount'] = np.log1p(df['Amount'])

# Dropping a column
df.drop('Time', axis=1, inplace=True)
```

```python

# --- NumPy Operations ---

# Basic array
arr = np.array([1, 2, 3, 4])
print("Array:", arr)
```

```python
# Mathematical operations
print("Mean:", np.mean(arr))
print("Standard deviation:", np.std(arr))
print("Max:", np.max(arr))
print("Min:", np.min(arr))
```

```python
# Random data
random_data = np.random.randn(1000)
```

```python

# --- Matplotlib Operations ---

# Basic line plot
plt.plot(arr)
plt.title('Line Plot')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()
```

```python
# Histogram of Amounts
plt.hist(df['Amount'], bins=50, color='skyblue')
plt.title('Transaction Amounts')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()
```

```python

# --- Seaborn Operations ---

# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), cmap='coolwarm', annot=False)
plt.title('Correlation Heatmap')
plt.show()

```

```python
# Distribution plot
sns.histplot(df['Amount'], bins=50, kde=True)
plt.title('Distribution of Amounts')
plt.show()
```

```python

# Boxplot of Amount by Class
sns.boxplot(x='Class', y='Amount', data=df)
plt.title('Transaction Amount by Class (Fraud/Not Fraud)')
plt.show()
```
